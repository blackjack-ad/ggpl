### Join of pointsets: esempio

- Fornisco un inseme di punti e il loro join e' il loro guscio convesso
- La seconda figura è il guscio convesso delle prime due
- Si ottiene unendo i punti
- MK trasforma un punto in uno 0-poliedro
- AA significa applied all. Ottengo sempre una lista
- con S=AA(FOIN)... prendo i primi quattro punti di P, dall'elemento 4 a 7 e poi il 7

```python
from pyplasm import *

pts = [[0,0],[.5,0],[0,.5],[.5,.5],[1,.5],[1.5,.5],[1.5,1],[.25,1]] # coords
P = AA(MK)(pts) # 0-polyhedra
S = AA(JOIN)([P[0:4],P[4:7],P[7]]) # array of d-polyhedra
H = JOIN(S) # 2-polyhedron
```

```python
VIEW(STRUCT(AA(SKELETON(1))(S)))
```

### Simplicial complex: esempio 1

```python
from larlib import * # import LAR library
V,CV = larSimplexGrid1([5,5,5]) # structured simplicial grid
FV = larSimplexFacets(CV) # 2-simplicial grid
EV = larSimplexFacets(FV) # 1-simplicial grid
VIEW(EXPLODE(1.5,1.5,1.5)(MKPOLS((V,CV))))
VIEW(EXPLODE(1.5,1.5,1.5)(MKPOLS((V,FV))))
VIEW(EXPLODE(1.5,1.5,1.5)(MKPOLS((V,EV))))
BV = [FV[t] for t in boundaryCells(CV,FV)] # boundary 2-simplices
VIEW(EXPLODE(1.5,1.5,1.5)(MKPOLS((V,BV))))
```

### Simplicial complex: esempio 2

- prima ho genrato una serie di punti random nel cerchio di raggio unitario
- con scipy.spatial ho importato Delaunay
- ho costruito un oggetto passando l'array dei vertici
- l'attributo "verticis" restituisce la lista dei triangoli dati come triple di vertici

```python
from larlib import *; from random import random as rand
points = [[2*PI*rand(),rand()] for k in range(1000)]
V = [[SQRT(r)*COS(alpha),SQRT(r)*SIN(alpha)] for alpha,r in points]
cells = [[k+1] for k,v in enumerate(V)]
VIEW(MKPOL([V,cells,None]))
from scipy.spatial import Delaunay
FV = Delaunay(array(V)).vertices
VIEW(EXPLODE(1.2,1.2,1)(MKPOLS((V,FV))))
VIEW(SKELETON(1)(STRUCT(MKPOLS((V,FV)))))
```
